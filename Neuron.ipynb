```python
#OR GATE 
#Using Threshold Function as the activation function
import random,numpy,math
weights = [random.random(), random.random(), random.random()]
bias = 1
lr = 1
sigmaJ = 10
def Neuron_Threshold(input1, input2, output):
    outputCalc = input1*weights[0] + input2*weights[1] + bias*weights[2]
    if outputCalc > 0:
        outputCalc = 1
    else:
        outputCalc = 0
    error = output - outputCalc
    weights[0] += error*input1*lr
    weights[1] += error*input2*lr
    weights[2] += error*bias*lr

def Predict_Threshold(input1, input2):
    outputCalc = input1*weights[0] + input2*weights[1] + bias*weights[2]
    if outputCalc > 0:
        outputCalc = 1
    else:
        outputCalc = 0
    return outputCalc

def Neuron_Sigmoid(input1, input2, output):
    outputCalc = input1*weights[0] + input2*weights[1] + bias*weights[2]
    outputCalc = 1/(1+numpy.exp(-outputCalc))
    error = output - outputCalc
    weights[0] += error*input1*lr
    weights[1] += error*input2*lr
    weights[2] += error*bias*lr

def Predict_Sigmoid(input1, input2):
    outputCalc = input1*weights[0] + input2*weights[1] + bias*weights[2]
    outputCalc = 1/(1+numpy.exp(-outputCalc))
    return outputCalc

def Neuron_Gaussian(input1, input2, output):
    outputCalc = input1*weights[0] + input2*weights[1] + bias*weights[2]
    outputCalc = numpy.exp(-(math.pow(outputCalc,2)/(2*math.pow(sigmaJ,2))))
    error = output - outputCalc
    weights[0] += error*input1*lr
    weights[1] += error*input2*lr
    weights[2] += error*bias*lr

def Predict_Gaussian(input1, input2):
    outputCalc = input1*weights[0] + input2*weights[1] + bias*weights[2]
    outputCalc = numpy.exp((-math.pow(outputCalc,2)/(2*math.pow(sigmaJ,2))))
    return outputCalc

def Main():
    ch = 0
    for i in range(50):
            Neuron_Threshold(0,0,0)
            Neuron_Threshold(1,0,1)
            Neuron_Threshold(0,1,1)
    for i in range(50):
            Neuron_Sigmoid(0,0,0)
            Neuron_Sigmoid(0,1,1)
            Neuron_Sigmoid(1,0,1)
    for i in range(50):
            Neuron_Gaussian(0,0,0)
            Neuron_Gaussian(0,1,1)
            Neuron_Gaussian(1,0,1)
    while ch!=4:
        print('\n1.Threshold Function\n2.Sigmoid Function\n3.Gaussian Function\n4.Exit')
        print('Enter your choice')
        ch = int(input())
        if ch!=4:
            print('Enter the value of x and y')
            x = int(input())
            y = int(input())
        if ch == 1:
            print('Predicted Output(Threshold) : ',Predict_Threshold(x,y))
        elif ch == 2:
            print('Predicted Output(Sigmoid) : ',Predict_Sigmoid(x,y))
        elif ch == 3:
            print('Predicted Output(Gaussian) : ',Predict_Gaussian(x,y))

if __name__ == '__main__':
    Main()
```
